**Description**: Demonstrates that subsampling / replicating within each dataset is
important. Data comes from running [`meta.py`](./meta.py). (`cd` to here before running
`python meta.py`.)

**Estimated runtime**: ~5 sec.

```python
from typing import Literal

import numpy as np
import polars as pl
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
sns.set_theme(style="darkgrid")
```

```python
zero_shot = True
if zero_shot:
    num_train: Literal[100] = 100
    num_test: Literal[100] = 100

    treatment, control = "test", "extra"
    color = "orange"  # blue for extra - base / pretraining boost

    desc = "evaluation bias (Mistral 7B, zero-shot)"
    title = f"$n={num_test}$, {desc}"

    path = f"meta_zero_shot_m{num_train}_n{num_test}_treatment.csv"
else:
    num_train: Literal[50, 100] = 100
    num_test: Literal[50, 100, 200, 500] = 500

    treatment, control = "test", "extra"
    color = "orange"  # blue for extra - base / pretraining boost

    desc = "evaluation bias"
    title = f"$m={num_train}, n={num_test}$, {desc}"

    path = f"meta_m{num_train}_n{num_test}_treatment.csv"
```

```python
df = pl.read_csv(path)
len(df)  # num simulated datasets
```

For the cheeky coin flip analogy, let's p-hack :-]

```python
magic_number = 0.04
df.select(
    (pl.col("mean") < -magic_number).or_(pl.col("mean") > magic_number)
).mean().item()
```

What percent of these are big differences have their 89% credible interval on one side
or the other?

```python
df.filter((pl.col("mean") < -magic_number).or_(pl.col("mean") > magic_number)).select(
    (pl.col("hdi_5.5%") > 0).xor(pl.col("hdi_94.5%") > 0).not_()
).mean().item()
```

```python
means_sorted = df["mean"].sort().to_numpy()
y1, y2 = np.searchsorted(means_sorted, [-magic_number, magic_number]) / len(df)
```

```python
xlim = (round(means_sorted[0], 2), round(means_sorted[-1], 2))
```

I want a CDF b/c it makes it easier to answer fun meta questions of the form: what's the
probability that we would've concluded `< insert wacky result >` had we not subsampled?

```python
ax: plt.Axes = sns.ecdfplot(df["mean"], color=color)
ax.set_xlim(xlim)
ax.set_xticks([xlim[0], -magic_number, 0, magic_number, xlim[1]])
_ = ax.set_title(title)
_ = ax.set_xlabel("posterior mean of $\\beta$")
_ = ax.set_ylabel("cumulative proportion")

_ = ax.fill_betweenx(
    np.arange(len(df)) / len(df),
    df.sort("mean")["hdi_5.5%"].to_numpy(),
    df.sort("mean")["hdi_94.5%"].to_numpy(),
    color=color,
    alpha=0.3,
    label="89% HDI",
)

ax.vlines(-magic_number, 0, y1, linestyle="dashed", color="gray")
ax.hlines(y1, xlim[0], -magic_number, linestyle="dashed", color="gray")

ax.vlines(magic_number, 0, y2, linestyle="dashed", color="gray")
ax.hlines(y2, xlim[0], magic_number, linestyle="dashed", color="gray")

plt.legend();
```
